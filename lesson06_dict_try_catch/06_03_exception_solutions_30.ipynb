### é¡Œç›® 13 è§£æï¼šä½¿ç”¨è€…è¼¸å…¥é©—è­‰è¿´åœˆ

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š
1. **while è¿´åœˆèˆ‡ç•°å¸¸è™•ç†çµåˆ**ï¼šé‡è©¦æ©Ÿåˆ¶
2. **è¨ˆæ•¸å™¨æ§åˆ¶**ï¼šé™åˆ¶é‡è©¦æ¬¡æ•¸
3. **break èªå¥**ï¼šæˆåŠŸå¾Œè·³å‡ºè¿´åœˆ

**ğŸ“ è§£é¡Œé—œéµ**ï¼š
- ä½¿ç”¨ while è¿´åœˆé…åˆè¨ˆæ•¸å™¨ attempts
- ValueError ç™¼ç”Ÿæ™‚å¢åŠ è¨ˆæ•¸å™¨
- æˆåŠŸè½‰æ›å¾Œä½¿ç”¨ break è·³å‡ºè¿´åœˆ
- é¡¯ç¤ºå‰©é¤˜é‡è©¦æ¬¡æ•¸çµ¦ç”¨æˆ¶

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- ç¬¬ 1 æ¬¡è¼¸å…¥ "abc" â†’ ç„¡æ•ˆï¼Œå‰©é¤˜ 2 æ¬¡
- ç¬¬ 2 æ¬¡è¼¸å…¥ "5.5" â†’ ç„¡æ•ˆï¼Œå‰©é¤˜ 1 æ¬¡
- ç¬¬ 3 æ¬¡è¼¸å…¥ "42" â†’ æˆåŠŸï¼Œè¿”å› 42

```python
# é¡Œç›® 13 å®Œæ•´è§£ç­”
user_inputs = ["abc", "5.5", "42"]
input_index = 0
max_retries = 3
attempts = 0

def mock_input():
    global input_index
    if input_index < len(user_inputs):
        val = user_inputs[input_index]
        input_index += 1
        return val
    return None

print("è«‹è¼¸å…¥ä¸€å€‹æ•´æ•¸ï¼ˆæœ€å¤š 3 æ¬¡æ©Ÿæœƒï¼‰")
while attempts < max_retries:
    try:
        user_input = mock_input()
        if user_input is None:
            break
        num = int(user_input)
        print(f"âœ… æ‚¨è¼¸å…¥çš„æ•¸å­—ï¼š{num}")
        break
    except ValueError:
        attempts += 1
        if attempts < max_retries:
            print(f"âŒ ç„¡æ•ˆè¼¸å…¥ï¼Œè«‹é‡è©¦ï¼ˆå‰©é¤˜ {max_retries - attempts} æ¬¡ï¼‰")
        else:
            print("âŒ å·²é”åˆ°æœ€å¤§é‡è©¦æ¬¡æ•¸")
```

### é¡Œç›® 14 è§£æï¼šå®‰å…¨è§£æè³‡æ–™æª”æ¡ˆ

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š
1. **CSV æ•¸æ“šè§£æ**ï¼šä½¿ç”¨ `split()` åˆ†å‰²å­—ç¬¦ä¸²
2. **æ•¸æ“šé©—è­‰**ï¼šæª¢æŸ¥æ¬„ä½æ•¸é‡å’Œé¡å‹
3. **è‡ªå®šç¾© ValueError**ï¼šæ‰‹å‹•æ‹‹å‡ºç•°å¸¸

**ğŸ“ è§£é¡Œé—œéµ**ï¼š
- ä½¿ç”¨ `split(",")` åˆ†å‰² CSV è¡Œ
- æª¢æŸ¥ `len(parts) != 3` æ‹‹å‡º ValueError
- `int()` è½‰æ›å¤±æ•—è‡ªå‹•æ‹‹å‡º ValueError
- æ‰€æœ‰ ValueError çµ±ä¸€è™•ç†ï¼Œè·³éç„¡æ•ˆè¡Œ

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- "Alice,25,50000" â†’ æˆåŠŸè§£æ
- "Bob,thirty,60000" â†’ ValueErrorï¼ˆå¹´é½¡ç„¡æ³•è½‰æ•´æ•¸ï¼‰
- "Charlie,35" â†’ ValueErrorï¼ˆæ¬„ä½ä¸è¶³ï¼‰
- "Diana,28,70000" â†’ æˆåŠŸè§£æ

```python
# é¡Œç›® 14 å®Œæ•´è§£ç­”

def parse_csv(data):
    """å®‰å…¨è§£æ CSV æ ¼å¼çš„è³‡æ–™"""
    result = []
    for line in data:
        try:
            # åˆ†å‰²è³‡æ–™
            parts = line.split(",")
            
            # æª¢æŸ¥æ¬„ä½æ•¸é‡
            if len(parts) != 3:
                raise ValueError("æ¬„ä½æ•¸é‡ä¸æ­£ç¢º")
            
            # è§£æå„æ¬„ä½
            name = parts[0]
            age = int(parts[1])  # å¯èƒ½æ‹‹å‡º ValueError
            salary = int(parts[2])  # å¯èƒ½æ‹‹å‡º ValueError
            
            # åŠ å…¥çµæœ
            result.append({"name": name, "age": age, "salary": salary})
            
        except ValueError as e:
            print(f"è­¦å‘Š: ç„¡æ³•è§£æ '{line}' - {e}")
    
    return result

# æ¸¬è©¦è³‡æ–™
csv_data = [
    "Alice,25,50000",
    "Bob,thirty,60000",    # å¹´é½¡ç„¡æ³•è½‰æ•´æ•¸
    "Charlie,35",          # æ¬„ä½ä¸è¶³
    "Diana,28,70000"
]

print("è§£æçµæœï¼š")
parsed_data = parse_csv(csv_data)
for person in parsed_data:
    print(person)
```

### é¡Œç›® 15 è§£æï¼šè¨ˆç®—æ©Ÿæ‡‰ç”¨

**ğŸ’¡ æ ¸å¿ƒæ¦‚å¿µ**ï¼š
1. **å­—å…¸é‹ç®—ç¬¦æ˜ å°„**ï¼šä½¿ç”¨ dict å„²å­˜é‹ç®—å‡½æ•¸
2. **KeyError è™•ç†**ï¼šæ•ç²ä¸æ”¯æ´çš„é‹ç®—ç¬¦
3. **ZeroDivisionError è™•ç†**ï¼šæ•ç²é™¤é›¶éŒ¯èª¤
4. **é‹ç®—ç¬¦æ¨¡å¼åŒ¹é…**ï¼šä½¿ç”¨å­—å…¸å–ä»£ if-elif

**ğŸ“ è§£é¡Œé—œéµ**ï¼š
- ä½¿ç”¨ `operations = {"+": lambda...}` æ˜ å°„é‹ç®—ç¬¦
- `operations[operator]` æŸ¥æ‰¾é‹ç®—ï¼Œä¸å­˜åœ¨æ™‚æ‹‹å‡º KeyError
- é™¤æ³•é‹ç®— `b = 0` æ™‚æ‹‹å‡º ZeroDivisionError
- åˆ†åˆ¥è™•ç†å…©ç¨®ç•°å¸¸ï¼Œçµ¦å‡ºæ˜ç¢ºéŒ¯èª¤è¨Šæ¯

**ğŸ” åŸ·è¡Œçµæœ**ï¼š
- calculate(10, 5, "+") â†’ 15
- calculate(10, 5, "/") â†’ 2.0
- calculate(10, 0, "/") â†’ ZeroDivisionErrorï¼ˆé™¤ä»¥é›¶ï¼‰
- calculate(10, 5, "%") â†’ KeyErrorï¼ˆä¸æ”¯æ´çš„é‹ç®—ç¬¦ï¼‰

```python
# é¡Œç›® 15 å®Œæ•´è§£ç­”

def calculate(a, b, operator):
    """åŸ·è¡ŒåŸºæœ¬é‹ç®—ä¸¦è™•ç†ç•°å¸¸"""
    # å®šç¾©é‹ç®—ç¬¦å­—å…¸
    operations = {
        "+": lambda x, y: x + y,
        "-": lambda x, y: x - y,
        "*": lambda x, y: x * y,
        "/": lambda x, y: x / y
    }
    
    try:
        # æŸ¥æ‰¾ä¸¦åŸ·è¡Œé‹ç®—
        operation = operations[operator]  # å¯èƒ½æ‹‹å‡º KeyError
        result = operation(a, b)          # å¯èƒ½æ‹‹å‡º ZeroDivisionError
        return result
        
    except KeyError:
        print(f"éŒ¯èª¤: ä¸æ”¯æ´çš„é‹ç®—ç¬¦ '{operator}'")
        print("æ”¯æ´çš„é‹ç®—ç¬¦: +, -, *, /")
        
    except ZeroDivisionError:
        print("éŒ¯èª¤: é™¤æ•¸ä¸èƒ½ç‚ºé›¶")

# æ¸¬è©¦
print("10 + 5 =", calculate(10, 5, "+"))
print("10 / 5 =", calculate(10, 5, "/"))
print("\næ¸¬è©¦ç•°å¸¸æƒ…æ³ï¼š")
calculate(10, 0, "/")  # ZeroDivisionError
calculate(10, 5, "%")  # KeyError
```
